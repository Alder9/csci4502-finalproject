```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
```

```python
death_path = 'pubg-match-deaths/deaths/kill_match_stats_final_0.csv'
agg_path = 'pubg-match-deaths/aggregate/agg_match_stats_0.csv'

pubg_death = pd.read_csv(death_path)
erangel_Map = pubg_death.loc[(pubg_death["map"]=="ERANGEL")]
bad_erangel_match_id = list(erangel_Map.loc[erangel_Map["killed_by"]=="Down and Out", "match_id"].drop_duplicates())
solo_erangel = erangel_Map.loc[~erangel_Map["match_id"].isin(bad_erangel_match_id)]

pubg_agg = pd.read_csv(agg_path)
solo_agg = pubg_agg.loc[~pubg_agg["match_id"].isin(bad_erangel_match_id)]
```

```python
erangel_ids = []
for _id in solo_agg["match_id"]:
    if _id in solo_erangel["match_id"]:
        erangel_ids.append(_id)
erangel_agg = solo_agg.loc[solo_agg["match_id"].isin(erangel_ids)]
erangel_agg.head()
```
