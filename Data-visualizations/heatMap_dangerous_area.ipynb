```python
import pandas as pd
import numpy as np

print('Loading data sets...')
dfK1 = pd.read_csv('deaths/kill_match_stats_final_1.csv', dtype = {'killed_by':str, 'killer_name':str, 'killer_placement':float, 'killer_position_x':float, 'killer_position_y':float, 'map':str, 'match_id':str, 'victim_name':str, 'victim_placement':float, 'victim_position_x':float, 'victim_position_y':float} ) #Load kill data set for 3rd kill data file
print('Player death and aggregate data loaded.\n')
```

```python
dfK1.head(100)
```

```python
import matplotlib.pyplot as plt
dfK1_copy = dfK1.loc[dfK1["map"] == "MIRAMAR" ]
dfK1_copy.dropna(how="any")
dfK1_copy.plot(kind = "scatter", x= "victim_position_x", y="victim_position_y",alpha=0.4)
plt.show()
```

```python
dfK1_copy.plot(kind="scatter", x="victim_position_x", y="victim_position_y",
    s=dfK1_copy['victim_placement']/100, label="victim",
    c="victim_placement", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.legend()
plt.show()
```

```python
#print(dfK1_copy[dfK1_copy['victim_placement'].isnull()])
dfK1_copy=dfK1_copy.dropna(subset=["victim_placement"])
```

```python
print(dfK1_copy[dfK1_copy['victim_placement'].isnull()])
```

```python
import matplotlib.image as mping
MIRAMAR_IMAGE = mping.imread('miramar.jpg')
ax = dfK1_copy.plot(kind="scatter", x="victim_position_x", y="victim_position_y",
    s=dfK1_copy['victim_placement']/10, label="victim",
    c="victim_placement", cmap=plt.get_cmap("jet"),
    colorbar=False, alpha=0.4, figsize=(15,10),
)
plt.imshow(MIRAMAR_IMAGE, extent=[0, 800000, 0, 800000],alpha=0.5)
plt.ylabel("victim_position_x", fontsize=14)
plt.xlabel("victim_position_y", fontsize=14)

victims = dfK1_copy["victim_placement"]
tick_values = np.linspace(victims.min(), victims.max(), 11)
cbar = plt.colorbar()
cbar.ax.set_yticklabels([v for v in tick_values], fontsize=14)
cbar.set_label('victims', fontsize=16)

plt.legend(fontsize=16)
plt.show()
```
